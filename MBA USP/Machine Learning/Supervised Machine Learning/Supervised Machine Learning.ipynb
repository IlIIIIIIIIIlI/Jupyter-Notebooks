# Supervised Machine Learning

Os Modelos Supervisionados em Machine Learning fazem predições acerca de um comportamento com base em um conjunto de dados. Essas predições são simplificações da realidade, e necessariamente omitem ou distorcem informações acerca da mesma.

Nos modelos GLM (Generalized Linear Model) ou Modelos Supervisionados, temos a variável dependente $Y$ que seria o outcome do modelo, e a variável $X_1$, $X_2$, $X_3$,...$X_k$ as variáveis explicativas (ou independentes como chamadas em algumas fontes). Temos que $Y=f(X_1, X_2, X_3, ..., X_k)$

Abaixo segue uma tabela sobre alguns modelos supervisionados:


\begin{array}{|c|c|c|}
\hline
\textbf{Modelo de Regressão} & \textbf{Característica da Variável Dependente} & \textbf{Distribuição} \\
\hline
\text{Linear} & \text{Quantitativa} & \text{Normal} \\
\text{Com transformação de Box-Cox} & \text{Quantitativa} & \text{Normal Após a Transformação}\\
\text{Logística Binária} & \text{Qualitativa com 2 Categorias (Dummy)} & \text{Bernoulli}\\
\text{Logística Multinomial} & \text{Qualitativa M (M > 2) Categorias} & \text{Binomial}\\
\text{Poisson} & \text{Quantitatica com Valores Inteiros e Não Negativos (Dados de Contagem)} & \text{Poisson}\\
\text{Binomial Negativo} & \text{Quantitativa com valores inteiros e não negativos (dados de contagem)} & \text{Poisson-Gama}\\
\hline
\end{array}

# 1 - Supervised Machine Learning: Regressão Linear Simples

O objetivo é desenvolver uma equação linear que apresente a relação entre uma variável dependente e uma variável explicativa.

Partindo da **Equação geral da reta:** 

\begin{equation}
Y - Y_0 = m(X - X_0)
\end{equation}

\begin{equation}
Y = Y_0 - m.X_0 + m.X
\end{equation}

Considerando $Y_0 - m.X_0 = \alpha$ e $m=\beta$, então temos:

$Y_i = \alpha + \beta.X_i + u$

Onde:

$Y_i=$ Valor da i-ésima observação da variável dependente para o i-ésimo dado
$\alpha = $ intercepto\
$\beta = $ coeficiente de inclinação \
$\alpha = $intercepto. Ponto que intercepta o eixo Y quando $x=0$\
$u = Y_i - \hat{Y}_i$ termo de erro onde $\hat{Y}_i$ é o valor previsto para o ponto $Y_i$


Nessa técnica, existe uma variável chamada de $R^2$ que é chamada de coeficiente de determinação (ou coeficiente de ajuste, explicação) que representa o peso que uma variável tem na variação de outra variável. Exemplo: a distância tem um coeficiente de determinação alto em relação a explicar o tempo que se leva para fazer um trajeto.

**Ajuste de uma reta aos pontos**

Existem algumas condições necessárias para se ajustar uma reta a um conjunto de pontos:

1. $\sum^{n}_{i=1} u_i = 0$
2. $min(\sum^{n}_{i=1} (u_i)^2)$

Esse é conhecido como o método dos *Mínimos Quadrados Ordinários* MQO

**O que é o $ R^2 $ na Análise de Regressão?**

O $ R^2 $, também conhecido como **coeficiente de determinação**, é uma medida estatística que indica a proporção da variância na variável dependente ($ Y $) que é explicada pela variável independente ($ X $) no modelo de regressão. Ele varia entre 0 e 1 e fornece uma indicação de quão bem os dados observados se ajustam ao modelo de regressão.

- **$ R^2 = 1 $**: Indica que o modelo de regressão explica perfeitamente a variação em $ Y $.
- **$ R^2 = 0 $**: Indica que o modelo de regressão não explica nenhuma variação em $ Y $.

**Como Interpretar o $ R^2 $**

- Um $ R^2 $ mais próximo de 1 indica que o modelo faz um bom trabalho em explicar a variação em $ Y $ com base em $ X $.
- Um $ R^2 $ próximo de 0 indica que o modelo não explica bem a variação em $ Y $, e outros fatores não incluídos no modelo podem estar influenciando $ Y $.

**Como Calcular o $ R^2 $?**

O $ R^2 $ é calculado com base nas seguintes fórmulas:

**1. **Resíduos ao Quadrado (SSE - Soma dos Quadrados dos Erros)**
A soma dos quadrados dos resíduos, $ SSE $, mede a variação que o modelo de regressão não consegue explicar:

$
SSE = \sum_{i=1}^{n} (Y_i - \hat{Y_i})^2
$

onde:
- $ Y_i $ é o valor observado da variável dependente.
- $ \hat{Y_i} $ é o valor predito pela equação de regressão.

**Soma Total dos Quadrados (SST - Total Sum of Squares)**
A soma total dos quadrados, $ SST $, mede a variação total em $ Y $ em relação à média de $ Y $:

$
SST = \sum_{i=1}^{n} (Y_i - \bar{Y})^2
$

onde:
- $ \bar{Y} $ é a média dos valores observados de $ Y $.

**Soma dos Quadrados Explicada (SSR - Regression Sum of Squares)**
A soma dos quadrados explicada, $ SSR $, mede a variação que o modelo de regressão consegue explicar:

$
SSR = \sum_{i=1}^{n} (\hat{Y_i} - \bar{Y})^2
$

**Fórmula para $ R^2 $**

O $ R^2 $ pode ser calculado de duas maneiras equivalentes:

1. Como a razão entre a soma dos quadrados explicada e a soma total dos quadrados:
$
R^2 = \frac{SSR}{SST}
$

2. Ou como 1 menos a razão entre a soma dos quadrados dos erros e a soma total dos quadrados:
$
R^2 = 1 - \frac{SSE}{SST}
$

---
**Exemplo de Cálculo do $ R^2 $**

Suponha que você tem os seguintes dados para $ Y_i $ (valores observados), $ \hat{Y_i} $ (valores preditos pelo modelo) e $ \bar{Y} $ (média de $ Y $):

- $ Y_i $ = [3, 5, 7, 9]
- $ \hat{Y_i} $ = [2.8, 5.1, 6.9, 9.2]
- $ \bar{Y} $ = 6

**Passo 1**: Calcular $ SSE $:

$
SSE = (3 - 2.8)^2 + (5 - 5.1)^2 + (7 - 6.9)^2 + (9 - 9.2)^2 = 0.04 + 0.01 + 0.01 + 0.04 = 0.1
$

**Passo 2**: Calcular $ SST $:

$
SST = (3 - 6)^2 + (5 - 6)^2 + (7 - 6)^2 + (9 - 6)^2 = 9 + 1 + 1 + 9 = 20
$

**Passo 3**: Calcular $ R^2 $:

$
R^2 = 1 - \frac{SSE}{SST} = 1 - \frac{0.1}{20} = 1 - 0.005 = 0.995
$

Neste exemplo, $R^2 = 0.995$, indicando que o modelo de regressão explica 99,5% da variação em $ Y $.

**Na análise de regressão simples** $R^2=\rho^2$ onde $\rho$ é o coeficiente de correlação de pearson.

**Conclusão**

O  $R^2$ é uma medida crucial para avaliar a qualidade do ajuste de um modelo de regressão. Ele indica a proporção da variação total em $Y$ que é explicada pela variável $X$ no modelo. Um valor de $R^2$ mais alto indica um modelo que melhor explica a variabilidade dos dados observados. Em resumo, quanto menor a dispersão dos dados em torno do ajuste, maior o valor de $R^2$.

---

Apenas o parâmetro $R^2$ não é suficiente para verificar se o ajuste tem significância estatística.

### 1.1.3 - Estimação do modelo de regressão linear simples

```python
# In[1.3]: Estimação do modelo de regressão linear simples

# Estimação do modelo
modelo = sm.OLS.from_formula('tempo ~ distancia', df_tempodist).fit()

# Observação dos parâmetros resultantes da estimação
modelo.summary()
```

### 1.1.4 - Avaliando o modelo de regressão

Aqui podemos ver alguns parâmetros importantes:

$\alpha=5.88$\
$\beta=1.42$\
$R^2=0.82$

Vale ressaltar que apenas os valores de $\alpha$, $\beta$ e $R^2$ **não são suficientes** para dizer se o ajuste é estatísticamente significativo. Para verificar se o ajuste é adequado, devemos olhar o resultado do **teste F**. Para isso olhamos a probabilidade da estatística F, se 

Prob (F-statistic) $<= 0.05$ então o ajuste é estatísticamente significativo. 

### 1.1.5 - Criando os plots da regressão

```python
# In[1.4]: Salvando fitted values (variável yhat) e residuals (variável erro)
#no dataset

df_tempodist['yhat'] = modelo.fittedvalues
df_tempodist['erro'] = modelo.resid
df_tempodist
```

```python
# In[1.5]: Gráfico didático para visualizar o conceito de R²

plt.figure(figsize=(15,10))
y = df_tempodist['tempo']
yhat = df_tempodist['yhat']
x = df_tempodist['distancia']
mean = np.full(x.shape[0] , y.mean(), dtype=int)

for i in range(len(x)-1):
    plt.plot(x, yhat, color='grey', linewidth=7)
    plt.plot([x[i], x[i]], [yhat[i], mean[i]], '--', color='darkorchid', linewidth=5)
    plt.plot([x[i], x[i]], [yhat[i], y[i]],':', color='limegreen', linewidth=5)
    plt.scatter(x, y, color='navy', s=220, alpha=0.2)
    plt.axhline(y = y.mean(), color = 'silver', linestyle = '-', linewidth=4)
    plt.title('R²: ' + str(round(modelo.rsquared, 4)), fontsize=30)
    plt.xlabel('Distância', fontsize=24)
    plt.ylabel('Tempo', fontsize=24)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.xlim(0, 35)
    plt.ylim(0, 60)
    plt.legend(['Fitted Values', 'Ychapéu - Ymédio', 'Erro = Y - Ychapéu'],
               fontsize=22, loc='upper left')
plt.show()
```

```python
# In[1.6]: Cálculo manual do R²

R2 = ((df_tempodist['yhat']-
       df_tempodist['tempo'].mean())**2).sum()/(((df_tempodist['yhat']-
                                        df_tempodist['tempo'].mean())**2).sum()+
                                        (df_tempodist['erro']**2).sum())

round(R2,4)
```

### 1.2.4 - Interpretação da estatística F de Fisher

Se:

- $F_{\text{calc}}>= F_{\text{crít}} \rightarrow H_1$ é aceito
    * p-value $<= 0,05$
    * pelo menos um valor de $\beta$ é estatísticamente significativo ($\neq 0$)

- $F_{\text{calc}} < F_{\text{crít}} \rightarrow H_0$ é aceito
    * p-value $> 0,05$
    * $\beta_1 = \beta_2 = ... = \beta_k = 0$
    * Não há modelo

Para saber quais valores de $\beta$ são estatísticamente significativos, precisamos utilizar o **teste t**.


#### 1.2.4.1 - Definição do p-value associado ao F calculado

```python
from scipy.stats import f

1 - f.cdf(F, df_modelo, df_residuos)
```

### 4.3.1 - Verificação da aderência dos resíduos à normalidade

Resíduos não aderentes à normalidade referem-se aos erros (ou resíduos) em um modelo estatístico que não seguem a distribuição normal. Em muitos métodos de análise estatística, como a regressão linear, uma suposição comum é que os resíduos do modelo (ou seja, a diferença entre os valores observados e os valores preditos pelo modelo) seguem uma distribuição normal, com média zero e variância constante. Quando os resíduos não aderem a essa suposição de normalidade, isso pode indicar que o modelo pode não ser o mais adequado ou que há outras questões a serem investigadas.

**Conceito de Resíduos**
Resíduos: São as diferenças entre os valores observados e os valores preditos por um modelo estatístico. Eles representam o erro de previsão do modelo para cada observação.
Importância da Normalidade dos Resíduos
Validação do Modelo: A suposição de normalidade dos resíduos é importante em muitas análises estatísticas, especialmente para testar hipóteses e calcular intervalos de confiança. Se os resíduos não forem normalmente distribuídos, os resultados desses testes podem ser inválidos.

**Diagnóstico de Modelos:** Verificar a normalidade dos resíduos é uma forma de diagnosticar a adequação de um modelo. Resíduos que não seguem a distribuição normal podem sugerir que o modelo não está capturando bem a estrutura dos dados.

**Como Identificar Resíduos Não Aderentes à Normalidade?**
Histograma dos Resíduos: Um histograma dos resíduos que não segue a forma de um sino (característico de uma distribuição normal) pode indicar que os resíduos não são normais.

- Gráfico Q-Q (Quantil-Quantil): Um gráfico Q-Q plota os quantis dos resíduos contra os quantis de uma distribuição normal teórica. Se os resíduos seguirem a normalidade, os pontos se alinharão aproximadamente em uma linha reta. Desvios significativos dessa linha sugerem que os resíduos não são normais.

- Testes de Normalidade: Testes estatísticos como o teste de Shapiro-Wilk ou o teste de Kolmogorov-Smirnov podem ser usados para avaliar formalmente se os resíduos seguem uma distribuição normal. Um p-valor pequeno nesses testes indica que é improvável que os resíduos sejam normais.

    * O teste de normalidade obedece o seguinte teste de hipótese: $H_0$: Não há diferença entre as distribuições a um determinado nível de significância, ou seja, segue uma distribuição normal (geralmente $p_\text{valor} > 0,05)$; naturalmente, $H_1$: Há diferença entre as distribuições, ou seja, a distribuição não é normal ($p_\text{valor}\leq 0,05$).

**Consequências dos Resíduos Não Aderentes à Normalidade**
Inferências Estatísticas Comprometidas: A normalidade dos resíduos é uma condição para a validade de muitos testes estatísticos. Se os resíduos não forem normais, as inferências, como testes de hipóteses e intervalos de confiança, podem ser menos confiáveis.

- Indicação de Modelagem Inadequada: Resíduos que não seguem a normalidade podem indicar que o modelo está incorretamente especificado. Pode haver variáveis importantes faltando, ou o relacionamento entre as variáveis pode não ser linear.

**O que Fazer se os Resíduos Não Aderirem à Normalidade?**
- Transformação dos Dados: Aplicar transformações nos dados (como a transformação logarítmica) pode ajudar a normalizar os resíduos.

- Modelos Alternativos: Considerar modelos estatísticos alternativos que não exijam a suposição de normalidade dos resíduos, como modelos de regressão robusta ou modelos não paramétricos.

- Incluir Variáveis Faltantes: Revisar o modelo para ver se variáveis importantes foram omitidas, o que poderia estar causando a falta de normalidade nos resíduos.

## 4.4 - Transformação de Box-Cox

A transformação de Box-Cox é uma normalização dos dados de tal forma, que a distribuição dos resíduos sejam aderentes à normalidade.

\begin{equation}
Y^*_\text{Box-Cox} = \frac{Y^\lambda - 1}{\lambda} = \alpha + \beta_1.X_1 + \beta_2.X_2 + ... + \beta_3.X_3
\end{equation}

Os diferentes valores de $\lambda$ determinam o tipo de transofrmação:

- $\lambda = -1 \longrightarrow$ inverso
- $\lambda = 0 \longrightarrow$ logarítmico $ln(Y)$
- $\lambda = 0,5 \longrightarrow$ raiz
- $\lambda = 1 \longrightarrow$ linear
- $\lambda = 2 \longrightarrow$ quadrático
- $\lambda = 3 \longrightarrow$ cúbico

Portanto:
\begin{equation}
Y^{*}_\text{Box-Cox} (\lambda) = 
\begin{cases}
\dfrac{Y^\lambda - 1}{\lambda} & \text{se } \lambda \neq 0 \\
\\
ln(Y) & \text{se } \lambda = 0
\end{cases}
\end{equation}
Ao implementar o método no python, o algorítmo verifica qual o valor de $\lambda$ que maximiza a aderência à normalidade.

A transformação de Box-Cox só pode ser aplicada a dados que são estritamente positivos ($y>0$) pois a transformação com $\lambda = 0$ envolve o cálculo do logaritmo.

O principal objetivo é melhorar a normalidade e a homocedasticidade dos dados, o que pode melhorar a precisão das inferências estatísticas.

## 5.8 - Verificando a adesão à normalidade dos resíduos do modelo não linear

```python
# In[5.13]: Verificando a normalidade dos resíduos do 'modelo_step_empresas_bc'

# Teste de Shapiro-Francia: interpretação
teste_sf = shapiro_francia(modelo_step_empresas_bc.resid) #criação do objeto 'teste_sf'
teste_sf = teste_sf.items() #retorna o grupo de pares de valores-chave no dicionário
method, statistics_W, statistics_z, p = teste_sf #definição dos elementos da lista (tupla)
print('Statistics W=%.5f, p-value=%.6f' % (statistics_W[1], p[1]))
alpha = 0.05 #nível de significância
if p[1] > alpha:
	print('Não se rejeita H0 - Distribuição aderente à normalidade')
else:
	print('Rejeita-se H0 - Distribuição não aderente à normalidade')
```

```python
# In[5.14]: Plotando os novos resíduos do 'modelo_step_empresas_bc' e
#acrescentando uma curva normal teórica para comparação entre as distribuições

from scipy.stats import norm

# Calcula os valores de ajuste da distribuição normal
(mu, sigma) = norm.fit(modelo_step_empresas_bc.resid)

# Gráfico propriamente dito
plt.figure(figsize=(15,10))
sns.histplot(modelo_step_empresas_bc.resid, bins=20, kde=True, stat="density",
             color='indigo', alpha=0.4)
plt.xlim(-0.5, 0.5)
x = np.linspace(-0.5, 0.5, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)
plt.xlabel('Resíduos do Modelo Box-Cox', fontsize=20)
plt.ylabel('Frequência', fontsize=20)
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)
plt.show()
```

```python
# Visualizando os dois fitted values dos modelos 'modelo_step_empresas' e
#'modelo_step_empresas_bc' no dataset
df_empresas[['empresa','retorno','yhat_step_empresas','yhat_step_empresas_bc']]

# In[5.19]: Ajustes dos modelos: valores previstos (fitted values) X valores reais

from scipy.optimize import curve_fit

def objective(x, a, b, c, d, e, f):
    return (a * x) + (b * x**2) + (c * x**3) + (d * x**4) + (e * x**5) + f

xdata = df_empresas['retorno']
ydata_linear = df_empresas['yhat_step_empresas']
ydata_bc = df_empresas['yhat_step_empresas_bc']

plt.figure(figsize=(17,10))

popt, _ = curve_fit(objective, xdata, ydata_linear)
a, b, c, d, e, f = popt
x_line = np.arange(min(xdata), max(xdata), 1)
y_line = objective(x_line, a, b, c, d, e, f)
plt.plot(x_line, y_line, '--', color='darkorange', linewidth=3)

popt, _ = curve_fit(objective, xdata, ydata_bc)
a, b, c, d, e, f = popt
x_line = np.arange(min(xdata), max(xdata), 1)
y_line = objective(x_line, a, b, c, d, e, f)
plt.plot(x_line, y_line, '--', color='indigo', linewidth=3)

plt.plot(xdata,xdata, color='gray', linestyle='-')
plt.scatter(xdata,ydata_linear, alpha=0.5, s=150, color='darkorange')
plt.scatter(xdata,ydata_bc, alpha=0.5, s=150, color='indigo')
plt.title('Dispersão e Fitted Values dos Modelos Linear e Box-Cox',
          fontsize=20)
plt.xlabel('Valores Reais de Retorno', fontsize=17)
plt.ylabel('Fitted Values', fontsize=17)
plt.legend(['Stepwise','Stepwise com Box-Cox','45º graus'], fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```
